```
########### 태크닉 13 : 데이터에 오류가 있는 상태로 집계해보자 ##############
```

```

# 데이터의 오류가 어느정도 집계에 영향을 미치는지 확인해보자
# 데이터의 정합성이 얼마나 중요한지 알아보자
# 매출 이력서에서 상품별로 월 매출 합계를 집계해보자.

uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])
uriage_data["purchase_month"] = uriage_data["purchase_date"].dt.strftime("%Y%m")
res = uriage_data.pivot_table(index="purchase_month", columns="item_name",
                              aggfunc = "size", fill_value=0)

res

# 1행과 2행에서 날짜를 년월 까지로 변환했다.
# 3행에서 새로축에 구입연월 가로축에 상품 의 건수로 집계
# 4행은 결과 표시
# 컬럼이 불필요하게 너무 많다. 
# 원래 26개 상품이 99개 상품으로 집계가 되었다.

```

```
res = uriage_data.pivot_table(index="purchase_month", columns="item_name",
                              values="item_price", aggfunc="sum", fill_value=0)

res

# 데이터에 오류가 있는 상태로 집계 및 분석을 하면 전혀 의미 없는 결과가 나오게 된다.
# 따라서 데이터 전처리가 중요함을 확인할 수 있다.
```

```
################# 태크닉 15 : 금액의 결측지를 수정하자 ####################
```

```
uriage_data.isnull().any(axis=0)

# 결측지에 따라서 현장에서 조사 후 수정하거나 부서원이 결측지를 채워 넣는 등의
# 여러가지 대응이 가능하다.

```

```
flg_is_null = uriage_data["item_price"].isnull()
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
  price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg),"item_price"].max()
  uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"]==trg)] = price
uriage_data.head()

# 1행에서 item_price 중에 결측치가 있는 곳을 조사한다. 그 결과를 flg_is_null에 저장
# flg_is_null을 이용해서 결측치가 있는 상품명 리스트를 작성한다
# list() 는 변수의 값을 리스트 형식을 변환, loc() 는 조건에 일치하는 데이터 추출
# 여기서는 조건이 '금액에 결측치가 있다.
# item_name은 조건과 일치하는 데이터중에서 어떤 컬럼을 가져올지 지정
# -> 결측치가 존재하는 상품명을 가져온다
# unique
```

```
uriage_data.isnull().any(axis=0)
```

```
for trg in list(uriage_data["item_name"].sort_values().unique()):

  print(trg+"의 최고가 : "+ str(uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].max()) + "의 최저가 : " + str (uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].min(skipna=False)))



# 1행에서 모든 상품에 대해 반복문 적용. 상품에 설정된 최대 금액과 최소 금액을 출력
# 결과로 모든 상품의 최대 금액과 최소 금액이 일치하는 것으로 봐서 성공적으로 금액을 수정한 것을 확인
# min(skipna=False) 에서 skipna는 NaN의 무시 여부를 설정한다. False로 지정했기 때문에 
# NaN이 존재할 경우 최소값이 NaN으로 표시된다.
```

```
###############태크닉 17 : 날짜오류를 수정하자 ################
```

```
flg_is_serial = kokyaku_data["등록일"].astype("str").str.isdigit()
flg_is_serial.sum()

#엑셀 데이터를 취급할때 주의 해야할 점은 '서식이 다른 데이터가 섞여있다'는 것이다
```

```
kokyaku_data["등록일"]
```

```
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial,"등록일"].astype("float"), unit="D") + pd.to_datetime("1900/01/01")

fromSerial

# to_timedelta() 함수를 이용하면 고유번호 숫자를 진짜 날짜로 변환한다.

```

```
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial,'등록일'])
fromString
```

```
kokyaku_data["등록일"] = pd.concat([fromSerial, fromString])
kokyaku_data

# 숫자를 날짜로 변경 데이터 fromSerial과 서식을 변경한 fromString concat으로 결합하고
#원래 등록일 필드에 대입하여 갱신하자

# quiz 등록일에서 년월만 추출해 필드명 등록연월
# 등록월별 고객이름의 개수
```

```

######## 태크닉 18 : 고객이름을 키로 두개의 데이터를 결합(조인)하자
```

```
# 정합성이 없는 데이터들은 서로 다른 열을 지정해서 결합
# left_on으로 좌측 right_on 으로 우측 데이터를 지정하고 how 결합 방법으로 옆으로 붙인다.
# uriage_data를 기준으로 kokyaku_data를 결합한다. 
join_data = pd.merge(uriage_data, kokyaku_data, left_on="customer_name", right_on="고객이름", how="left")
join_data = join_data.drop("customer_name", axis=1)
join_data 
#데이터 결합을 통해서 분석에 적합한 데이터가 되었다
#이것을 데이터 정제라고 함
```

```
########## 태크닉 19 : 정제한 데이터를 덤프하자 
```

```
# 목적은 데이터 가공/정제 가 아니라 '데이터의 분석'을 하는 것이다.
# 깨끗해진 정제한 데이터를 파일로 새로 만들어 (데이터 덤프) 해두고
# 분석을 할때마다 정제된 파일을 읽어서 분석한다.

dump_data = join_data[["purchase_date","purchase_month","item_name","item_price","고객이름","지역","등록일"]]
dump_data

# 사소한 부분이지만 실제 현장에서도 큰 데이터를 다룰 때 고생하지 않기 위해서
# 평소에 이런 습관을 길러두는 것이 좋을것으로 사료된다.
```

```
dump_data.to_csv("dump_data.csv", index=False)
```
